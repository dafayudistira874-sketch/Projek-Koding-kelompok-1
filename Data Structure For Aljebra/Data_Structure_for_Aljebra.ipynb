# **8.Norms and Unit Vector**

```python
#buat vector x dengan ukuran 3x1
import numpy as np
x = np.array([[84],
             [79],
             [89]])
x
```

```python
#hitung nilai L2 norm dari vector x
np.linalg.norm(x)
```

```python
#buat vector v dengan elemen (3,4)
v = np.array([[84],[89],[90],[67],
              [75],[78],[73],[71],
              [68],[64],[69],[67]])
v
```

```python
#hitung nilai L2 norm dari vector v
np.linalg.norm(v)
```

```python
#tampilkan unit vector dari vector v
unit_vector = v / np.linalg.norm(v)
unit_vector
```

```python
#hitung panjang unit vector dari vector v
len = np.linalg.norm(v)
len
```

# **9.Basis, Orthogonal, and Orthonormal Vectors**

```python
#buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
i = np.array([1,0])
i
```

```python
j = np.array([0,1])
j
```

```python
#buktikan bahwa vector i dan vector j itu orthogonal
np.dot(i, j)
```

```python
#Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1,2])
w = np.array([2,-1])
np.dot(v, w)
```
