#**16.Tensor Reduction**

```python
import numpy as np
```

```python
#buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[5, 25, 50], [61, 2, 15]])
X
```

```python
#hitung berapa hasil penjumlahan elemen matrix X dengan fungsi sum
X.sum()
```

```python
#hitung hasil penjumlahan matrix X berdasarkan row nya
X.sum(axis=1)
```

```python
#hitung hasil penjumlahan matrix X berdasarkan kolom nya
X.sum(axis=0)
```

```python
#hitung nilai minimal dari matrix X
min_all = np.min(X)
min_all
```

```python
#hitung nilai maximal dari matrix X
max_all = np.max(X)
max_all
```

```python
#hitung nilai rata-rata dari matrix X
mean_all = np.mean(X)
mean_all
```

```python
#hitung hasil perkalian seluruh elemen matrix X
X.prod()
```

##Menggunakan TensorFlow

```python
import tensorflow as tf
```

```python
#buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[5, 25, 50], [61, 2, 15]])
X_tf
```

```python
#hitung hasil penjumlahan seluruh elemen matrix X
np.sum(X)
```

```python
#hitung hasil penjumlahan seluruh elemen matrix X_tf
tf.reduce_sum(X_tf)
```

```python
#hitung hasil penjumlahan matrix X_tf berdasarkan row nya
tf.reduce_sum(X_tf, 1)
```

```python
#hitung hasil penjumlahan matrix X_tf berdasarkan kolom nya
tf.reduce_sum(X_tf, 0)
```

```python
#hitung nilai minimal dari martix X_tf
min_all = tf.reduce_min(X_tf)
min_all
```

```python
#hitung nilai maximal dari matrix X_tf
max_all = tf.reduce_max(X_tf)
max_all
```

```python
#hitung nilai rata-rata dari matrix X_tf
mean_all = tf.reduce_mean(X_tf)
mean_all
```

```python
#hitung hasil perkalian seluruh elemen matrix X_tf
tf.reduce_prod(X)
```

# **21.Visualizing Linear Systems**


1.buat visualisasi dari persamaan berikut:

y = 2x+1

3x-y = 5


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 10, 400)

y1 = 2 * x + 1
y2 = 3 * x - 5

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = 2x + 1', color='blue', linewidth=2)
plt.plot(x, y2, label='3x - y = 5 (y = 3x - 5)', color='red', linewidth=2)

plt.scatter(6, 13, color='black', zorder=5)
plt.text(6.2, 12, 'Titik Potong (6, 13)', fontsize=10)

plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True, which='both', linestyle='--')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.title('Visualisasi Sistem Persamaan Linear')
plt.legend()
plt.xlim(-1, 10)
plt.ylim(-2, 15)
```
