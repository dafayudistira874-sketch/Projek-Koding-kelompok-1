# **23.frobenius norm**

```python
#1. menggunakan numpy, pytoch dan tensorflow buatlah matrix ukuran 3x3 lalu hitung nilai frobenius norm
import numpy as np
X = np.array([[1,2,3],[4,5,6],[7,8,67]])
X
```

```python
np.linalg.norm(X)
```

```python
import torch
x_pt = torch.tensor([[1,2,3],[4,5,6],[7,8,6]], dtype=float)
x_pt
```

```python
torch.norm(x_pt)
```

```python
import tensorflow as tf
x_tf = tf.Variable([[1,2,3],[4,5,6],[7,8,67]], dtype=float)
x_tf
```

```python
tf.norm(x_tf)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import tensorflow as tf
```

```python
#Dengan Numpy
A = np.array([[4,2], [-5,-3]])
B = np.array([[4,-7]])
A
A_inv = np.linalg.inv(A)
C = A_inv.dot(B.T)
C
```

```python
#Dengan Pytorch
x = torch.tensor([[4.0, 2.0],[-5.0, -3.0]] )

y = torch.tensor([[4.0], [-7.0]])
x_inv = torch.linalg.inv(x)

solution = x_inv @ y

print(solution)
```

```python
#dengan Tensorflow
a = tf.Variable([[4.0, 2.0],[-5.0, -3.0]] )
b = tf.Variable([[4.0], [-7.0]])
a_inv = tf.linalg.inv(a)
solution = a_inv @ b
print(solution)
```
